```python
from rac.experiment_data import ExperimentReader
metrics = ["rand", "ami", "time", "time_select_batch", "time_update_clustering", "num_violations"]
ac = ExperimentReader(metrics=metrics)

config = {
    "general_options": {
        "experiment_name": "tmlr_rebuttal_synthetic_new",
        "num_repeats": 14,
        "n_workers": 14,
        "local": True,
        "verbose": False
    },
    "experiment_options": {
        "seed": 33,
        "num_feedback": 0.001,
        "noise_level": [0.4],
        "persistent_noise_level": 0.0,
        "save_matrix_data": False,
        "infer_sims": False,
        "predict_sims": False,
        "clustering_alg": "CC",
        "running_avg": True,
        "num_maxmin_edges": [0, 0.005, 0.05, 0.2, 0.6, 1],
    },
    "sim_init_options": {
        "K_init": 10,
        "sim_init": [0.1],
        "sim_init_type": ["random_clustering"]
    },
    "query_strategy_options": {
        "acq_fn": ["maxexp"],
        "eps": 0.3,
        "beta": 1,
        "tau": [7],
        "alpha": 1
    },
    "dataset_options": {
        "dataset": "synthetic",
        "n_samples": 500,
        "n_clusters": 10,
        "class_balance": None
    }
}

data = ac.read_all_data(folder="../experiment_results_local/tmlr_rebuttal_synthetic_new/")
ac.generate_AL_curves(
    data,
    save_location="../plots/tmlr_rebuttal_synthetic_new/MAXEXP2",
    categorize=["noise_level", "sim_init_type", "sim_init"],
    compare=["acq_fn", "num_maxmin_edges"],
    vary=["x"],
    auc=True,
    summary_method="auc_max_ind",
    indices=[], 
    threshold=1,
    err_style="band",
    marker="o",
    markersize=6,
    capsize=6,
    linestyle="solid",
    **config
)
```
